**Funções Básicas**

```python
# Create a new DataFrame with only the duplicate rows
duplicate_rows = df.loc[df.duplicated()]
duplicate_rows
```

```python
#### 6- Find and drop duplicate rows based on InvoiceNo, StockCode, Quantity, and UnitPrice columns

df_without_duplicate_orders = df.drop_duplicates(subset=["InvoiceNo", 
                                "StockCode","Quantity", "UnitPrice"])
df_without_duplicate_orders
```

```python
# Drop duplicates while keeping the first non-NaN value based on InvoiceNo, StockCode, and CustomerID columns

df_keep_first = df.drop_duplicates(subset=["InvoiceNo", "StockCode","CustomerID"], keep="first")
df_keep_first
```

```python
#TAREFA: Drop duplicates while keeping the last order based on StockCode and InvoiceWeekday columns

# Convert InvoiceDate column to datetime
df_weekday = df

df_weekday['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Create a new column for weekday
df_weekday['InvoiceWeekday'] = df['InvoiceDate'].dt.weekday

# Drop duplicates based on StockCode and InvoiceWeekday
df_unique_stock_day = df_weekday.drop_duplicates(subset=["StockCode", "InvoiceWeekday"], keep="last")
df_unique_stock_day
```

```python
# TAREFA: Drop all duplicate invoices as it reflects to multiple products in the same invoice
# OBS: DEfault parametro keep = first
df_black_friday = df.drop_duplicates(subset="InvoiceNo", keep=False)
df_black_friday
```

```python
# TAREFA: Drop duplicate countries while keeping the first row

df_unique_countries = df.drop_duplicates(subset="Country")
df_unique_countries.head()
```

```python
#TAREFA:  Drop duplicate products while keeping last based on StockCode, Description, and UnitPrice

df_unique_products = df.drop_duplicates(
subset=["StockCode", "Description", "UnitPrice"], keep="last")

df_unique_products
```

```python
#TAREFA - Drop all duplicate rows based on TotalCost and CustomerID while keeping first

df_cost = df
# Create a new column for TotalCost
df_cost['TotalCost'] = df_cost['Quantity']*df_cost['UnitPrice']

df_customer_unique_payments = df_cost.drop_duplicates(
subset=["TotalCost", "CustomerID"])

df_customer_unique_payments
```

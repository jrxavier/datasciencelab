```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dados/Pokemon.csv")
```

```python
df.head()
```

```python
df.info()
```

***Distribuição de tipos de pokemon***

```python
df["Type 1"].value_counts().plot(kind="pie", autopct="%1.1f%%", cmap="tab20c", figsize=(10,8));
```

***Distribuição dos totais dos Pokemons***

```python
df["Total"].plot(kind="hist", figsize=(8,6));
```

```python
df["Total"].plot(kind='box', vert=False, figsize=(8,6));
```

***Distribuição de Pokemons Legendários***

```python
df["Legendary"].value_counts().plot(kind='pie', autopct="%1.1f%%", cmap="Set3", figsize=(8,6))
```

***Filtros Básicos***

****Quantos Pokemons existem com o valor de ATTACK maior do que 150?****

```python
df["Attack"].plot(kind='box', vert=False);
```

```python
df.loc[df["Attack"] > 150].shape
```

```python
df.query("Attack > 150")
```

****Selecionar apenas os Pokemons com Speed de 10 ou menos****

```python
sns.boxplot(data=df, x="Speed")
```

```python
slow_pokemon_df = df.loc[df["Speed"] <= 10]
slow_pokemon_df
```

```python
#Quantos Pokemons têm valor de Sp. Def de 25 ou menos?

(df["Sp. Def"] <= 25).sum()
```

```python
# Selecionar todos os Pokemons Não São Legendários !!

legendary_df = df.loc[~ df["Legendary"]]
legendary_df
```

```python
# Identifique um Oulier

ax = sns.scatterplot(data=df, x="Defense", y="Attack")
ax.annotate("Esse ponto", xy=(228,10), xytext=(150,10), color='red', arrowprops=dict(arrowstyle="->", color="red"))
```

```python
#Identificando o ponto acima - Pokemon Shuckle

df.sort_values(by=['Defense', 'Attack'], ascending=[False, True]).head()
```

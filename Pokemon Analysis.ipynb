```python
# Quantos Pokemons Fire-Flying existem?

(
    (df['Type 1'] == 'Fire') & 
    (df['Type 2'] == 'Flying')
).sum()
```

```python
df.loc[(
    (df['Type 1'] == 'Fire') & 
    (df['Type 2'] == 'Flying')
)]
```

```python
df.query("`Type 1`== 'Fire' and `Type 2`== 'Flying'")
```

```python
# How many 'Poison' Pokemons are across both types?

(
    (
        (df['Type 1'] == 'Poison') | 
        (df['Type 2'] == 'Poison')
    )
).sum()
```

```python
df.query("`Type 1`== 'Poison' or `Type 2`== 'Poison'").shape
```

```python
# What Pokemon of Type 1 Ice has the strongest defence

df.loc[df['Type 1'] == 'Ice'].sort_values('Defense', ascending=False).head(1)
```

```python
#What's the most common type of Legendary Pokemons?

df.loc[df['Legendary'], 'Type 1'].value_counts().plot(kind='bar');
```

```python
# What's the most powerful Pokemon from the first 3 generations of type Water?

df.loc[
    (df['Type 1'] == 'Water') &
    (df['Generation'].isin([1,2,3]))
].sort_values('Total', ascending=False).head()
```

```python
# What's the most powerful (by Total) Dragon (Type 1 and Type 2) from the last two generations?

df.loc[
    (   (df['Type 1'] == 'Dragon') |
        (df['Type 2'] == 'Dragon') 
    ) 
    &
    (df['Generation'].isin({5,6}))
    
].sort_values('Total', ascending=False).head()
```

```python
# Select the most powerful Fire-type Pokemons (Attack value above 100 and Type 1 equals to Fire)

df.loc[
    (   (df['Type 1'] == 'Fire') &
        (df['Attack'] > 100) 
    ) 
].head()
```

```python
powerful_pokemon_df = df.query("Attack > 100 and `Type 1` == 'Fire'")
powerful_pokemon_df.head()
```

```python
# Select all Water-type, Flying type Pokemons (Type 1 = Water and Type 2 = Flying)

df.query("`Type 1` == 'Water' and `Type 2`=='Flying'").head()
```

```python
# Select specific columns of Legendary Pokemons of type Fire: Select only Name, Attack and Generation
# Syntax

# df.loc [
#  (CONDITIONS FOR INDEX), (COLUMNS)
# ]

df.loc[
    (
        (df["Type 1"] == 'Fire') & 
        df['Legendary']
    ), 
    ["Name", "Attack", "Generation"]
]
```

```python
# Select SLOW and FAST Pokemon

ax = df['Speed'].plot(kind='hist', figsize=(10,5), bins=100);
ax.axvline(df['Speed'].quantile(0.05), color='red')
ax.axvline(df['Speed'].quantile(0.95), color='red')
```

```python
boton_5 = df['Speed'].quantile(0.05)
top_5 = df['Speed'].quantile(0.95)
(boton_5, top_5)
```

```python
df.loc[
    (df['Speed'] < boton_5) |
    (df['Speed'] > top_5)
]
```

```python
df.query("Speed < @boton_5 or Speed > @top_5").sort_values("Speed")
```

```python
# Find the Ultra Powerful Legendary Pokemon

fig, ax = plt.subplots(figsize=(14,7))
sns.scatterplot(data=df, x="Defense", y="Attack", hue='Legendary', ax=ax)
```
